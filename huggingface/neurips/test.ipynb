```python
import pandas as pd
df = pd.read_csv("neurips_2025_papers_with_institutions.csv")
```

```python
import pandas as pd

# 1. Explode the institutions column on ';'
df_exp = (
    df.assign(institution=df['institutions'].str.split(';'))
      .explode('institution')
)

# 2. Normalize case and strip whitespace
df_exp['institution_norm'] = df_exp['institution'].str.lower().str.strip()

# 3. Filter where institution contains "google"
df_google = df_exp[df_exp['institution_norm'].str.contains('google', na=False)]
```
