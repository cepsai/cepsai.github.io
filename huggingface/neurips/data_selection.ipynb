```python
keys = [
    "NeurIPS 2025 Datasets and Benchmarks Track oral",
    "NeurIPS 2025 Datasets and Benchmarks Track spotlight",
    "NeurIPS 2025 spotlight",
    "NeurIPS 2025 oral",
]

filtered = (
    df["venues"]
      .apply(lambda d: {k: d.get(k) for k in keys} if isinstance(d, dict) else {})
)

out = df.join(pd.DataFrame(filtered.tolist(), index=df.index)).fillna({k: 0 for k in keys})

```

```python
out[['name','parent','urban_zone','country','parent2']+keys].rename(columns={"parent": "area", "parent2": "region"}).to_csv("neurips_spotlight_oral_2025.csv",index=0)
```
